```python
import re
from textblob import  TextBlob
```

```python
def clean_tweet(tweet):
    return(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)",
                          " ",tweet).split()))
```

```python
def get_tweet_sentiment(tweet):
    analysis=TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity>0:
        return ('Positive')
    elif analysis.sentiment.polarity==0:
        return ('Neutral')
    else:
        return('Negative')
```

```python
records_all=db_tweets.find()
sentiments_total={'Positive':0,'Neutral':0,'Negative':0}
for i in records_all:
    sentiment=get_tweet_sentiment(i['text'])
    print(sentiment,"===>",i['text'])
    sentiments_total[sentiment]=sentiments_total[sentiment]+1
    print('--------------------------')

    
print(sentiments_total)
    
    
```

```python
from matplotlib import pyplot as plt
slices = [sentiments_total['Positive'],sentiments_total['Neutral'],
          sentiments_total['Negative']]
activities = ['Positive','Neutral','Negative']
cols = ['c','m','r']

plt.pie(slices,
        labels=activities,
        colors=cols,
        shadow= True,
        autopct='%1.1f%%')

plt.title('Sentiment Analysis of Tweets')
plt.legend()
plt.show()
```
